Pandas code on Student Analysis Project


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#create Dataset 
data = {
    "Name" : ["A","B","C","D","E"],
    "Marks" : [80,90,89,98,None],
    "City" : ["Delhi","Mumbai","Noida","Delhi","Muzaffarnagar"]
}
df = pd.DataFrame(data)
print("Original Data :\n",df)
```

```python
#Basic Functions 
print("\nFirst 5 rows:\n", df.head())
print("\nLast 5 rows:\n",df.tail())
print("\nData Info:\n")
print(df.info())
print("\nDescription:\n",df.describe())
```

```python
#Data Cleaning
print("\nChecking Null Values:\n",df.isnull())
#Fill missing values
df["Marks"]=df["Marks"].fillna(df["Marks"].mean())
print("\nAfter Filling Missing Values:\n",df)
```

```python
#Remove duplicates (if any)
df = df.drop_duplicates()
print("df",df)
```

```python
#Data Analysis
print("\nAverage Marks:",df["Marks"].mean())
print("Maximum Marks:",df["Marks"].max())
print("Minimum Marks:",df["Marks"].min())
```

```python
#Data Visualization 
df.plot(x="Name" , y="Marks",kind="bar",title="Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```
